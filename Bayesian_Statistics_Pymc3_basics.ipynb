### 1a Below function genreates sample_set from Bernouli distribution, with input parameter and sample size set. 

```
def return_sampleset(par_p,size):
  return np.random.binomial(size=size, n=1, p= par_p)
```

### 4000 size sample set for p=0.75  

```
y_org_data=return_sampleset(0.75,4000)  
```

```

datapoints=[1,2,3,4,1000,4000]
priors = [[1,1],[2,3],[4,4],[3,2]]
prtitle= ['Non_informative','weak peak at 0.25','peak at 0.5','weak peak at 0.75']

fig, ax = plt.subplots(4,len(datapoints))

for p in range(len(priors)):
  for  i in range(len(datapoints)) :
    with pm.Model() as model:
      # define the prior
      theta = pm.Beta('theta', priors[p][0], priors[p][1])  # prior
      # define the likelihood
      y = pm.Bernoulli('y', p=theta, observed=y_org_data[0:datapoints[i]])
      # Generate a MCMC chain
      trace = pm.sample(500)
    title = prtitle[p]+'datapoints- '+str(datapoints[i])  
    #ax[p, i].hist(trace['theta'], color='skyblue')
    pm.plot_posterior(trace['theta'], ax=ax[p,i], color='skyblue');
    ax[p, i].set_title(title) 
    ax[p, i].set_xlim(0, 1)
    ax[p, i].set_xlabel(r'$\theta$')
    #ax[i].set_ylabel('Position in Chain')  

fig.set_figheight(13)
fig.set_figwidth(18)
fig.tight_layout() 
plt.show()      
```

### 1c plots of posterior distrbution for following priors, with 1,2,3, 4 datapoints. 4 data items one-by-one versus taking data points all at once.

```
datapoints=y_org_data[0:4]
priors = [[1,1],[2,3],[4,4],[3,2]]
prtitle= ['Non_informative','weak peak at 0.25','peak at 0.5','weak peak at 0.75']

fig, ax = plt.subplots(4,len(datapoints)+1)

# taking data points one at a time
for p in range(len(priors)):
  pa = priors[p][0]
  pb = priors[p][1]
  for  i in range(len(datapoints)) :
    with pm.Model() as model:
      # define the prior
      theta = pm.Beta('theta', pa, pb)  # prior
      # define the likelihood
      y = pm.Bernoulli('y', p=theta, observed=datapoints[i])
      # Generate a MCMC chain
      trace = pm.sample(500)
    if datapoints[i]==1:
      pa+=1
    else:
      pb+=1    
    title = prtitle[p]+'datapoint no- '+str(i+1)  
    #ax[p, i].hist(trace['theta'], color='skyblue')
    pm.plot_posterior(trace['theta'], ax=ax[p,i], color='skyblue',hdi_prob=0.95);
    ax[p, i].set_title(title) 
    ax[p, i].set_xlim(0, 1)
    ax[p, i].set_xlabel(r'$\theta$')
    #ax[i].set_ylabel('Position in Chain') 

# taking data points all  at once
for p in range(len(priors)):
  with pm.Model() as model:
      # define the prior
      theta = pm.Beta('theta', priors[p][0], priors[p][1])  # prior
      # define the likelihood
      y = pm.Bernoulli('y', p=theta, observed=datapoints)
      # Generate a MCMC chain
      trace = pm.sample(500)
  title = prtitle[p]+'4 datapoints'+str(i+1)  
  pm.plot_posterior(trace['theta'], ax=ax[p,4], color='red',hdi_prob=0.95);
  ax[p, 4].set_title(title) 
  ax[p, 4].set_xlim(0, 1)
  ax[p, 4].set_xlabel(r'$\theta$')




fig.set_figheight(13)
fig.set_figwidth(18)
fig.tight_layout() 
plt.show()      
```

(b) For two priors given below compute the posterior distributions for the three parameters for sample sizes
of 10, 100 and 1000 and plot the distribution and the 95% HDI.

are the sample mean and sample standard deviation of the ten-sized sample.






## A simple description for ht-wt is a linear dependence wË = Î²1h + Î²2 where Î²1, Î²0 are the two parameters of the formula and Ëw is the predicted value of the weight and h is the height.(src : class notes).


h â¼ N (5, 0.5)
$\beta$1  = 0.75
$\beta$2  = 15
$\sigma$   = 1
refrence -https://docs.pymc.io/notebooks/getting_started.html


```
def generate_hw(no_of_datapnt):
  beta1=0.75
  beta2=15
  sig= 1
  hights = np.random.normal(5,0.5,no_of_datapnt)
  wights = []
  for i in hights:
    temp_wt = i*beta1+beta2
    wights.append(np.random.normal(temp_wt,sig,1)[0])
  return hights,wights  
    
```

# Weak prior

```
datax,datay=generate_hw(1000)
master_x,master_y=datax,datay
fig, ax = plt.subplots(3, 3)
for i in range(3):
  reg1_model = pm.Model()
  with reg1_model:
    # Priors for unknown model parameters
    beta1 = pm.Uniform("beta1",lower = -100,upper=100, shape=1 )
    beta2 = pm.Uniform("beta2", lower=-100, upper = 100, shape=1)
    sigma = pm.Uniform("sigma",lower =0, upper = 100)
    #sigma = pm.HalfNormal("sigma", sigma=1)
    # Expected value of outcome
    mu = beta2+ beta1 * datax[0:10**(i+1)] 
    # Likelihood (sampling distribution) of observations
    Y_obs = pm.Normal("Y_obs", mu=mu, sigma=sigma, observed=datay[0:10**(i+1)])
    trace = pm.sample(500, return_inferencedata=False)
    #display(az.summary(trace, round_to=2))
  plotvar = trace.varnames[3:]
  count = 0
  for j in plotvar:
    pm.plot_posterior(trace[j], ax=ax[i, count], color='skyblue',hdi_prob=0.95)
    title = 'Non informative:'+str(10**(i+1))+'-datapoints'
    ax[i, count].set_title(title) 
    #ax[i, count].set_xlim(0, 1)
    ax[i, count].set_xlabel(j)
    count+=1


fig.set_figheight(10)
fig.set_figwidth(10)
fig.tight_layout() 
plt.show()     
#fig, ax = plt.subplots(2, 3)
 #pm.plot_posterior(theta_sample, ax=ax[0, 0], color='skyblue');
```
